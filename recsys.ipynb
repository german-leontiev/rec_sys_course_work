```python
import pandas as pd
import numpy as np
```

```python
df_prod = pd.read_csv("products.csv")
```

```python
df_trans = pd.read_csv("transactions.csv")
```

```python
df_trans.fillna(-1, inplace=True)
```

```python
mappers = {"encode":{
    "department": {v:k for k, v in enumerate(df_prod.department.drop_duplicates().values)},
    "aisle": {v:k for k, v in enumerate(df_prod.aisle.drop_duplicates().values)}
},
          "decode":{
    "department": {k:v for k, v in enumerate(df_prod.department.drop_duplicates().values)},
    "aisle": {k:v for k, v in enumerate(df_prod.aisle.drop_duplicates().values)}
          }}
```

```python
df_prod.department = df_prod.department.apply(lambda x: mappers["encode"]["department"][x])
```

```python
df_prod.aisle = df_prod.aisle.apply(lambda x: mappers["encode"]["aisle"][x])
```

```python
df_prod.drop(columns=["product_name"], inplace=True)
```

```python
df = df_trans.merge(df_prod)
```
